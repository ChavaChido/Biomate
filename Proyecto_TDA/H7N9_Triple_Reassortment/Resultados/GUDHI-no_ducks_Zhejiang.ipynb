```python
import argparse
import numpy as np
import gudhi
import matplotlib.pyplot as plot
```

```python
sG3 = [
    [0., 1221., 1393., 1335., 1394., 318., 1266., 310., 1271.],
    [1221., 0., 611., 637., 614., 1231., 442., 1226., 460.],
    [1393., 611., 0., 555., 9., 1413., 383., 1408., 421.],
    [1335., 637., 555., 0., 558., 1344., 626., 1341., 653.],
    [1394., 614., 9., 558., 0., 1416., 386., 1411., 424.],
    [318., 1231., 1413., 1344., 1416., 0., 1102., 19., 1094.],
    [1266., 442., 383., 626., 386., 1102., 0., 1098., 93.],
    [310., 1226., 1408., 1341., 1411., 19., 1098., 0., 1082.],
    [1271., 460., 421., 653., 424., 1094., 93., 1082., 0.]
]

# Convertir la lista a una matriz NumPy
sG3 = np.array(sG3)
```

```python
max_edge_length = 4000
max_dimension = 5
band = 0.0

print("#####################################################################")
print("RipsComplex creation from defined distance matrix G3")
   
rips_complex = gudhi.RipsComplex(
   distance_matrix=sG3, max_edge_length=max_edge_length  
)
```

```python
simplex_tree = rips_complex.create_simplex_tree(max_dimension=max_dimension)
   
message = "Number of simplices=" + repr(simplex_tree.num_simplices()) 
print(message)
  
diag = simplex_tree.persistence()

print("betti_numbers()=")
print(simplex_tree.betti_numbers())
```

```python
gudhi.plot_persistence_diagram(diag, band=band)
plot.show()
```
